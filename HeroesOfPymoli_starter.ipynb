```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data_df = pd.read_csv(file_to_load)
```

```python
#purchase_data.value_counts()

total_players = len(purchase_data_df["SN"].unique())

summary_table=pd.DataFrame({"Total Players" : total_players},index =[0])

#summary_table.set_Index([pd.Index([1]),'Total_Players'])
summary_table

```

```python

    #remove duplicates from SN column
#newpurchase_data=purchase_data_df.drop_duplicates(subset ="SN", keep = 'first',inplace= True)

# Calculate the number of unique items in the DataFrame  
uniqe_count = len(purchase_data_df["Item ID"].unique())
avgprice = purchase_data_df["Price"].mean()
noofpurchases=len(purchase_data_df["Purchase ID"])
total_revenue =  purchase_data_df["Price"].sum()
```

```python
# Place all of the data found into a summary DataFrame
format_dict = {'Average Price':'${0:,.0f}', 'Total Revenue': '${0:,.2f}'}
summary_table = pd.DataFrame({"Number Of unique Items": [uniqe_count],
                              "Average Price": avgprice,
                              "Number Of Purchases": noofpurchases,
                              "Total Revenue": total_revenue}).style.format(format_dict)


summary_table
```

```python

#remove duplicates from SN column
newpurchase_data = purchase_data_df.copy()

newpurchase_data.drop_duplicates(subset ="SN", keep = 'first',inplace= True)

newpurchase_data.head()
#type(newpurchase_data)
```

```python
gender_counts= newpurchase_data["Gender"].value_counts()

 


# Convert the gender_counts Series into a DataFrame
gender_counts_df = pd.DataFrame(gender_counts)


gender_counts_df = gender_counts_df.rename(
    columns={"Gender": "Total Count"})


gender_counts_df['Percentage Of Players']= round((gender_counts_df["Total Count"]/total_players)*100 ,2)

gender_counts_df["Percentage Of Players"]= gender_counts_df["Percentage Of Players"].map("{:.2f}%".format)
gender_counts_df.head()
```

```python


gendergroup_data_df = purchase_data_df.groupby('Gender')


totalpurchase_price = round(gendergroup_data_df["Price"].sum(),2)
avg_purchaseprice =round(gendergroup_data_df["Price"].mean(),2)


purchase_count = gendergroup_data_df["Price"].count()

# Calculate Normalized Purchasing (Average Total Purchase per Person)
normalized_total = round(totalpurchase_price / gender_counts_df["Total Count"],2)

gender_demographics = pd.DataFrame({"Purchase Count": purchase_count, "Average Purchase Price": avg_purchaseprice,"Total Purchase Value":
                                   totalpurchase_price,"AVG Total Purchase per person" : normalized_total})


gender_demographics["Average Purchase Price"] = gender_demographics["Average Purchase Price"].map("${:.2f}".format)
gender_demographics["Total Purchase Value"] = gender_demographics["Total Purchase Value"].map("${:,}".format)
gender_demographics["AVG Total Purchase per person"] = gender_demographics["AVG Total Purchase per person"].map("${:.2f}".format)
gender_demographics.head()

#namegroup_data = purchase_data.groupby("SN")
#totalpriceperperson=namegroup_data["Price"].mean()

#gender_demographics_totals = player_demographics["Gender"].value_counts()
#gender_demographics_percents = gender_demographics_totals / num_players
#gender_demographics = 

```

```python

#newpurchase_data = purchase_data_df.copy()

#newpurchase_data.drop_duplicates(subset ="SN", keep = 'first',inplace= True)

bins = [0,9.90,14.90,19.90,24.90, 29.90,34.90,39.90,9999]

# Create the names for the  bins
group_names = [ "<10", "10-14", "15-19"," 20-24","25-29","30-34","35-39","40+"]


newpurchase_data['age_ranges'] = pd.cut(newpurchase_data["Age"], bins, labels = group_names)


groupby_agebins = newpurchase_data.groupby('age_ranges')
total_count = groupby_agebins["SN"].count()
percentage_of_players =(round( groupby_agebins["SN"].count() / total_players,4))*100

age_demographics = pd.DataFrame({"Total Count": total_count, "Percentage of Players": percentage_of_players})

age_demographics["Percentage of Players"] = age_demographics["Percentage of Players"].map("{:.2f}%".format)
age_demographics
```

```python
#purchasing analysis by age

purchase_data_df['age_ranges'] = pd.cut(purchase_data_df["Age"], bins, labels = group_names)



purchase_coun = purchase_data_df.groupby(["age_ranges"]).count()["Price"].rename("Purchase Count")
totalpurchase_pric = round(purchase_data_df.groupby(["age_ranges"]).sum()["Price"].rename("Total purchase value"),2)
avg_purchasepric = purchase_data_df.groupby(["age_ranges"]).mean()["Price"].rename("Average Purchase Price")


#purchase_coun = groupby_agebins["Price"].count()

# Calculate Normalized Purchasing (Average Total Purchase per Person)
normal_total = round(totalpurchase_pric /age_demographics["Total Count"],2)

purchase_demographics = pd.DataFrame({"Purchase Count": purchase_coun, "Average Purchase Price": avg_purchasepric,"Total Purchase Value":
                                   totalpurchase_pric,"AVG Total Purchase per person" : normal_total})


purchase_demographics["Average Purchase Price"] = purchase_demographics["Average Purchase Price"].map("${:.2f}".format)
purchase_demographics["Total Purchase Value"] = purchase_demographics["Total Purchase Value"].map("${:,}".format)
purchase_demographics["AVG Total Purchase per person"] = purchase_demographics["AVG Total Purchase per person"].map("${:.2f}".format)
purchase_demographics.head(20)
```

```python
#group by sn
#groupby_sn = purchase_data_df.groupby("SN")

#Purchase_count = groupby_sn["Price"].count()

user_total =  round(purchase_data_df.groupby(["SN"]).sum()["Price"].rename("Total Purchase Value"),2)
user_average =round(purchase_data_df.groupby(["SN"]).mean()["Price"].rename("Average Purchase Price"),2)
user_count = purchase_data_df.groupby(["SN"]).count()["Price"].rename("Purchase Count")

#Average_Purchase_Price =  round(groupby_sn["Price"].mean(),2)

#total_purchase_value = round(groupby_sn["Price"].sum(),2)
topspender_df = pd.DataFrame({"Purchase Count": user_count, "Average Purchase Price": user_average,"Total Purchase Value":
                                   user_total})

topspender_df = topspender_df.sort_values("Total Purchase Value", ascending= False)
 
topspender_df["Average Purchase Price"]  = topspender_df["Average Purchase Price"].map("${:,}".format)
topspender_df["Total Purchase Value"]  = topspender_df["Total Purchase Value"].map("${:,}".format)
                             
topspender_df.head()
```
